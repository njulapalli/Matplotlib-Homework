```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from matplotlib.lines import Line2D
from matplotlib.patches import Circle

# File to Load (Remember to change these)
Pyber_city_data_to_load = "data/city_data.csv"
Pyber_ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
Pyber_city_df = pd.read_csv("data/city_data.csv")
Pyber_city_df.head(2)

Pyber_ride_df = pd.read_csv("data/ride_data.csv")
Pyber_ride_df.head(2)



```

```python
# Combine the data into a single dataset
Pyber_merge_df = pd.merge(Pyber_ride_df, Pyber_city_df, how="left", on=["city", "city"])
# Combine the data into a single dataset
Pyber_merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
#df_filtered = Pyber_merge_df[(Pyber_merge_df.type == 'Urban')]
#print(df_filtered)
#Pyber_merge_df.to_csv('merge.csv')
Pyber_plot_df = Pyber_merge_df.groupby(['city', 'type']).agg({'fare' : 'mean', 'ride_id':'count', 'driver_count' :'min'})
Pyber_plot_df.reset_index(level=['city', 'type'], inplace =True)

Pyber_plot_df.loc[Pyber_plot_df.type == 'Urban', 'color'] = 'Coral' 
Pyber_plot_df.loc[Pyber_plot_df.type == 'Suburban', 'color'] = 'skyblue'
Pyber_plot_df.loc[Pyber_plot_df.type == 'Rural', 'color'] = 'Gold'
Pyber_plot_df
```

```python
#Pyber_merge_df['ride_id']
plt.scatter(Pyber_plot_df['ride_id'], Pyber_plot_df['fare'], Pyber_plot_df['driver_count']*10, c= Pyber_plot_df['color'], alpha=0.5, edgecolors="black")
plt.xlim(0,42)
plt.ylim(19,44)
plt.title('Pyber Ride Sharing Data (2016)')
#plt.subtitle('Note: Circle size correlates with driver count per city')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides (Per City)')
plt.legend(['Urban', 'Suburban', 'Rural'])

#example from https://matplotlib.org/gallery/text_labels_and_annotations/custom_legends.html
legend_elements = [Line2D([0], [0], marker='o', color='black', label='Urban',
                          markerfacecolor='gold', markersize=10),
                   Line2D([0], [0], marker='o', color='black', label='Suburban',
                          markerfacecolor='skyblue', markersize=10),
                   Line2D([0], [0], marker='o', color='black', label='Rural',
                          markerfacecolor='Coral', markersize=10),
                   ]

plt.legend(handles=legend_elements, loc='upper right', title='City Types')
plt.show()
```
